## Linear Regression

Linear regression is a statistical method used to model the relationship between a dependent variable and one or more independent variables by fitting a straight line to the data. This line, known as the regression line, is represented by the equation:
$$y = mx + c$$

In this equation:

- $y$ is dependent variable (predicted outcome)
- $x$ is independent variable
- $m$ is the slope or gradient or **rate of change**
- $c$ is the y-intercept

#### Prediction price of a pizza of unknown size

| *Basic formula of regression* |
|:--:|
|Regression equation: $$y=mx+c$$ |
|Gradient or slope **(Fluctuation of Y due to the change of 1 unit of X)**: $$m= ∑ Product of Deviation/ Deviation of X $$|


Now lets predict the price -

- Price of pizza will be $$y=mx + c$$

- Gradiant or Slope
$$m= ∑ Product of Deviation/ Deviation of X $$
$$m= 12/ 8 $$
$$m= 1.5 $$

***that means if size (X) changes 1 inch then the price (Y) will be changed 1.5$***

- Intercept of Y
$$c= y- mx$$
$$c= Mean of Y - (m * mean of X)$$
$$c= 13 - (1.5 * 10)$$
$$c= 13-15$$
$$c= -2$$

***So mathematically the intercection point of the line is y=-2. As we know in y axis the x=0 so we can say that if the size(Y) = -2 then the price of the pizza(X)= 0***
